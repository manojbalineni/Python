### Loops


```python
user_input = ''
while(user_input != 'hello'):
  user_input = input("Enter String :")
  print(user_input)
```

```python
user_input = ''
while(user_input != 'hello'):
  user_input = input("Enter String :")
  print(user_input)
print("End of while loop")
```

```python
count = 0
while(count <= 5):
  print(count)
  count = count + 1
```

```python
names = ["manoj","swapna", "vissu"]
for name in names:
  name = name.upper()
  print(name)
```

```python
names = ["manoj","swapna", "vissu"]
for name in names:
  if name == 'swapna':
    continue
  name = name.upper()
  print(name)
```

```python
names = ["manoj","swapna", "vissu"]
for name in names:
  if name == 'swapna':
    break
  name = name.upper()
  print(name)
```

```python
count = 0
while True:
  if(count == 10):
    break
  print(count)
  count += 1
 
```
