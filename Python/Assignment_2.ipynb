1. The Smart Attendance Logger

The Goal: Manage a classroom attendance list and ensure no duplicate entries are recorded if a student taps their ID
card twice.

Concepts: set , function , while .

The Problem: Create a function mark_attendance() that keeps asking for student names until the user
types "STOP". Store names in a Set to automatically prevent duplicates.

Input: Multiple strings (names) entered one by one.

Output: The final count of unique students and the set itself.



```python
def mark_attendance():
  student_name = ""
  students_set = set()
  while(student_name != "STOP"):
    student_name = input("Enter Student name :")
    if(student_name != "STOP"):
      students_set.add(student_name)
  print(students_set)
  print(len(students_set))
    
```

```python
mark_attendance()
```

2. The Restaurant Bill Splitter

The Goal: Calculate how much each person owes, including a tax and a custom tip percentage.

Concepts: functions , default values , list .

The Problem: Write a function split_bill(items_prices, people, tax=0.05, tip=0.10) .
items_prices will be a List of decimals (floats).

Input: A list like [25.0, 15.0, 10.0] and an integer for people .

Output: A single float (the amount per person).


```python
def split_bill(items_prices , people , tax = 0.05 , tip = 0.10):
  total_bill = sum(items_prices)
  total_bill = total_bill + total_bill * tax
  total_bill =  total_bill + total_bill * tip
  split_per_person = total_bill / people
  return split_per_person
```

```python
split_per_person = split_bill([25.0, 15.0, 10.0], 5)
print(split_per_person)
```

3. The Digital Wardrobe Organizer

The Goal: Check if a specific outfit is available or if parts of it are missing.

Concepts: multiple if , list , input .

The Problem: You have a list wardrobe = ["Blue Jeans", "White Shirt", "Black Jacket"] . Ask the
user for 3 items they want to wear. Check each item individually against the list.

Input: Three separate input() strings (e.g., "Red Tie", "White Shirt", "Socks").

Output: A "Missing" warning for every item not found in the list.

```python
wardrobe =  ["Blue Jeans", "White Shirt", "Black Jacket"]
items_want_wear = input("Enter 3 items sepetated by comma :")
items = items_want_wear.split(",")
print(items)
for i in items:
  if i not in wardrobe:
    print("Missing item :" + i)
```

4. The E-Commerce Discount Engine

The Goal: Apply different discount rates based on the total purchase amount.

Concepts: if-elif-else , functions .
The Problem: Create a function apply_discount(total) .
If total > $500: 20% off.
If total > $200: 10% off.
Else: No discount.

Input: A single number (the total bill).
Output: The final price after the discount is subtracted.

```python
def apply_discount(total):
  if total > 500:
    discount = 0.2 * total
  elif(total > 200):
    discount = 0.1 * total
  else:
    discount = 0
  return total - discount
```

```python
apply_discount(520)
```

6. The Travel Itinerary Builder

The Goal: Print a numbered list of travel destinations for a user.

Concepts: for loop, list , enumerate .

The Problem: Given a list cities = ["Tokyo", "Paris", "New York", "London"] , print them as a
numbered list.

Input: None (Uses the hardcoded list).

Output: A vertically formatted list (e.g., "1. Tokyo").


```python
cities = ["Tokyo", "Paris", "New York", "London"]
answer_list = []
for index , city in enumerate(cities, start = 1):
  answer_list.append(str(index) + ". " + city)

print(answer_list)
```

7. The ATM Pin Validator

The Goal: Lock the user out after 3 incorrect attempts.

Concepts: while , if-else , break .

The Problem: Set correct_pin = "1234" . Allow the user 3 tries to enter the right PIN.

Input: Up to 3 string inputs.

Output: "Welcome" on success, or "Card Blocked" after the 3rd fail.


```python
correct_pin = "1234"
count = 0
while True:
  user_pin = input("Enter Pin :")
  if user_pin == correct_pin:
    print("Success")
    break
  else:
    count += 1
    if(count >= 3):
      print("In correct Pin and Card Blocked")
      break
    
    print("Pin In correct enter again ")
   
```

8. The Movie Ticket Age Checker

The Goal: Categorize a movie-goer and assign a price.

Concepts: tuple , if-elif-else .

The Problem: Use a Tuple to store fixed prices: PRICES = (0, 10, 15) (Free, Child, Adult). Ask for age
and determine which index from the tuple to use.

Input: An integer (Age).

Output: The price fetched from the tuple.

```python
prices = (0, 10, 15)
age = int(input("Enter age : "))
price = 0
if(age < 5):
  price = prices[0]
elif(age >= 5 and age < 15):
  price = prices[1]
else:
  price = prices[2]
print(price)
```

9. The Grocery Inventory Update

The Goal: Update the quantity of items in a store after a shipment arrives.

Concepts: dict , for loop.

The Problem: Given inventory = {"Apples": 50, "Bananas": 20} , a shipment of 10 more of each
arrives. Update the dictionary using a loop.

Input: None.

Output: The updated dictionary.


```python
inventory = {"Apples": 50, "Bananas": 20}
for i in inventory:
  inventory[i] = inventory[i] + 10

print(inventory)
```
