### Question - 1

The Office Door Badge You are programming a smart door for a tech office. The door should only open if the
user provides the correct "Access Code."

Input: User enters a 4-digit code.
Scenario: If the code is 1234 , print "Access Granted. Door Unlocked." Otherwise, print "Access Denied.
Security Alerted."

```python
door_code = int(input("Enter the 4-digit security code"))
if(door_code == 1234):
  print("Access Granted. Door Unlocked")
else:
  print("Access Denied. Security Alerted.")
```

### Question - 4

The Pizza Topping Calculator A customer is choosing toppings. Each topping adds a specific cost regardless of
the other toppings chosen.

Input: Three boolean (True/False) inputs for: extra_cheese , pepperoni , and olives .

Scenario: Initialize a total_bill = 10 .
If extra_cheese is True add 2 to bill.
If pepperoni is True add 3 to bill.
If olives is True add 1 to bill.
Output: Print the final total bill

```python
total_bill = 10
extra_cheese = False
pepperoni = True
olives = True

if extra_cheese:
  total_bill += 2
if pepperoni:
  total_bill += 3
if olives:
  total_bill += 1


print("Final Bill Amount is " , total_bill)
```

### Question - 5
The Traffic Light Sensor A self-driving car needs to interpret a single traffic light color.

Input: User enters "Red", "Yellow", or "Green".

Scenario: * If "Red" "Full Stop."
If "Yellow" "Slow down and prepare to stop."
If "Green" "Proceed with caution."
Anything else "Error: Signal Malfunction!"


```python
signal = input("Enter Signal :")
if signal == "Red":
  print("Full Stop")
elif signal == "Yellow":
  print("Slow down and prepare to stop")
elif signal == "Green":
  print("Proceed with caution")
else:
  print("Error: Signal Malfunction!!")
```

### Question - 6

Movie Theater Ticket Pricing A theater charges based on the age of the customer.
Input: User enters their age.

Scenario:
Age < 5 "Ticket is Free."
Age 5 to 17 "Child Ticket: $10."
Age 18 to 60 "Adult Ticket: $18."
Age > 60 "Senior Ticket: $12."

```python
age = int(input("Enter Age : "))
if (age < 5):
  print("Ticket is Free")
elif( age >= 5 and age <= 17):
  print("Child Ticket : $10")
elif ( age >= 18 and age <= 60):
  print("Adult Tikcet : $18")
else:
  print("Senior Ticket : $12")
```

### Question - 7

The 7-Day Steps Tracker A fitness app wants to show your activity for the week.
  
Data: A list of steps: [4000, 7500, 10000, 2000, 8000, 12000, 5000] .

Scenario: Loop through the list and for each day, print: "Day [X]: You walked [Steps] steps."

```python
steps = [4000, 7500, 10000, 2000, 8000, 12000, 5000]
for i in range(0, len(steps)):
  print("Day [" + str(i + 1) + "]: You walked [" + str(steps[i]) + "] steps.")
```
