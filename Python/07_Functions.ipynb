```python
a = int(input("Enter First number : "))
b = int(input("Enter Second number :"))
sum = a + b
print(sum)
```

```python
def sum_of_numbers():
  num_one = float(input("Enter first number :"))
  num_two = float(input("Enter Second number :"))
  total = num_one + num_two
  print(total)
```

```python
sum_of_numbers()
```

```python
def clean_string():
  a = input("Enter String")
  print(a)
  a = a.strip().lower()
  print(a)
```

```python
clean_string()
```

```python
count = 0
while count < 5:
  clean_string()
  count += 1
```

```python
def product_of_numbers():
  a = int(input("Enter first number : "))
  b = int(input("Enter second number :"))
  return a * b
```

```python
print(product_of_numbers())
```

```python
def subtract_two_numbers():
  a = int(input("Enter number"))
  b = int(input("Enter number :"))
  c = a - b
  return c
```

```python
result = subtract_two_numbers() + subtract_two_numbers()
print(result)
```
